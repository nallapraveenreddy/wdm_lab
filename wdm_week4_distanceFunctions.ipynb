#Implement the distance functions for assessing similarity between documents while taking care of standardizing the attributes

#Find 1.Jaccard Similarity, 2.Cosine similarity

```python
doc1 = "Data is the new oil of the digital economy where it is more useful in many applications"
doc2 = "Data is a new oil of economy"
```

#1.Jaccard Similarity

```python
def jaccard_similarity(doc1,doc2):
  words_doc1 = set(doc1.lower().split())
  words_doc2 = set(doc2.lower().split())
  #intersection
  intersection = words_doc1.intersection(words_doc2)
  #union
  union = words_doc1.union(words_doc2)
  return float(len(intersection)/len(union))
```

```python
j_s = jaccard_similarity(doc1,doc2)
j_s
```

```python
d1 = "The weather is so cool today and snow falling"
d2 = "weather is full of snow falling"
jaccard_similarity(d1,d2)
```

#2.Cosine Similarity

```python
data = [doc1,doc2]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vectorizer = CountVectorizer()
vector_matrix = count_vectorizer.fit_transform(data)
vector_matrix

```

```python
vector_matrix.toarray()
```

```python
import pandas as pd
def create_df(matrix,tokens):
  doc_names = [f'doc_{i+1}' for i,_ in enumerate(matrix)]
  df = pd.DataFrame(data=matrix, index=doc_names,columns=tokens)
  return(df)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity_matrix = cosine_similarity(vector_matrix)
create_df(cosine_similarity_matrix,['doc1','doc2'])
```
